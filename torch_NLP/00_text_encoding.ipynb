```python


import nltk
import seaborn as sns
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download('punkt')  # C:\Users\xx\AppData\Roaming\nltk_data
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))  #英文停用分词集合
stop_words.add('.')  #添加停用词
# stop_words.remove('an')

corpus = ['Time flies like an arrow.',
          'Fruit flies like a banana.']

word_tokens = word_tokenize(' '.join(corpus))
filtered_sentence = list(set(word_tokens) - stop_words)
print(filtered_sentence)

one_hot_vectorizer = CountVectorizer(binary=True)
one_hot = one_hot_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(one_hot, annot=True,
            cbar=False, xticklabels=filtered_sentence,
            yticklabels=['Sentence 2'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf = tfidf_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(tfidf, annot=True, cbar=False, xticklabels=filtered_sentence,
            yticklabels=['Sentence 1', 'Sentence 2'])
```
